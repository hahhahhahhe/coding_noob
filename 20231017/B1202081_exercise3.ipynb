```python
for i in range(1,10):
  for h in range(1,10):
    if i*h < 10:
      print('{} × {} = {}'.format(i,h,(i*h)),end = ('   '))
    elif i*h >= 10:
      print('{} × {} = {}'.format(i,h,(i*h)),end = ('  '))
  print('')
```

```python
i = 1
h = 1
while i < 10:
  while h < 10:
    if h != 9 and i*h < 10:
      print('{} × {} = {}'.format(h,i,(i*h)),end = ('   '))
    elif h != 9 and i*h >= 10:
      print('{} × {} = {}'.format(h,i,(i*h)),end = ('  '))
    elif h == 9:
      print('{} × {} = {}'.format(h,i,(i*h)))
    h = h + 1
  h = 1
  i = i + 1


```
